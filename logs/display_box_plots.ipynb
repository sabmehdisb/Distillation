```python
import json

import json
dataset="compas_0_optimised"
myfile = dataset+".json"

with open(myfile, 'r') as file_json:
    data_read = json.load(file_json)
```

```python
import statistics
accuracy_after_retraining_for_each_rule = data_read["accuracy_after_retraining_for_each_rule"]
accuracy_after_rectification_for_each_rule = data_read["accuracy_after_rectification_for_each_rule"]
number_of_nodes_after_retraining_for_each_rule = data_read["number_of_nodes_after_retraining_for_each_rule"]
number_of_nodes_after_rectification_for_each_rule = data_read["number_of_nodes_after_rectification_for_each_rule"]
depth_after_rectification_for_each_rule = data_read["depth_after_rectification_for_each_rule"]
depth_after_retraining_for_each_rule = data_read["depth_after_retraining_for_each_rule"]
accuracy_for_each_tree_on_test_set = data_read["accuracy_for_each_tree_on_test_set"]
number_of_different_instances_rectification = data_read["number_of_different_instances_rectification"]
number_of_different_instances_retraining = data_read["number_of_different_instances_retraining"]
accuracy_for_each_tree_on_test_set = data_read["accuracy_for_each_tree_on_test_set"]
columns_of_the_binarized_data_set = data_read["columns_of_the_binarized_data_set"]
time_rectification=data_read["time_rectification_for_each_rule"]
time_retrainig=data_read["time_retraining_for_each_rule"]
nbinstajoutéalltree=data_read["nbinstajoutéalltree_for_each_rule"]
nbinstsuppriméesalltree=data_read["nbinstsuppriméesalltree_for_each_rule"]
nbinstance_initial=data_read["nbinstance_initial"]
nbinstance_final=data_read["nbinstance_final"]
mediane = statistics.median(accuracy_for_each_tree_on_test_set)
print("Médiane :", mediane)
```

```python
# Retrieving data from the read data
import statistics
# Repetition of 'add' and 'ee' import, presumably from a library

# Rectification
# Create lists to store the medians for each position
median_size,median_precision,median_depth,median_num_different_instances,number_of_corrections_for_each_tree = [],[],[],[],[]

# Calculating the median for the length of instances
for i in number_of_different_instances_rectification:
    number_of_corrections_for_each_tree.append(len(i))
median_number_of_corrections_for_all_trees= int(statistics.median(number_of_corrections_for_each_tree))

# Calculating the median for each position
for i in range(max(len(sublist) for sublist in number_of_nodes_after_rectification_for_each_rule)):
    position_values = [sublist[i] for sublist in number_of_nodes_after_rectification_for_each_rule if i < len(sublist)]
    median_value = int(statistics.median(position_values))
    median_size.append(median_value)

# Calculating the median for each position
for i in range(max(len(sublist) for sublist in accuracy_after_rectification_for_each_rule)):
    position_values = [sublist[i] for sublist in accuracy_after_rectification_for_each_rule if i < len(sublist)]
    median_value = statistics.median(position_values)
    median_precision.append(median_value)

# Calculating the median for each position
for i in range(max(len(sublist) for sublist in depth_after_rectification_for_each_rule)):
    position_values = [sublist[i] for sublist in depth_after_rectification_for_each_rule if i < len(sublist)]
    median_value = int(statistics.median(position_values)) -1 #To exclude the root from the depth calculation.
    median_depth.append(median_value)

# Calculating the median for each position
for i in range(max(len(sublist) for sublist in number_of_different_instances_rectification)):
    position_values = [sublist[i] for sublist in number_of_different_instances_rectification if i < len(sublist)]
    median_value = int(statistics.median(position_values))
    median_num_different_instances.append(median_value)
temps_cumuile_rectification = [sum(sublist) for sublist in time_rectification]
temps_cumuile_rentrainement = [sum(sublist) for sublist in time_retrainig]
average_rectification = sum(temps_cumuile_rectification) / len(temps_cumuile_rectification)
average_retaining=sum(temps_cumuile_rentrainement) / len(temps_cumuile_rentrainement)
print("median_precision rectification",median_precision)
print("median_size rectification",median_size)
print("median_depth rectification",median_depth)
print("num_different_instances rectification", median_num_different_instances)

print("median_number_of_corrections_for_all_trees",median_number_of_corrections_for_all_trees)
#print("number_of_corrections_for_each_tree",number_of_corrections_for_each_tree)
print("average_rectification et temps_cumuile_rectification",average_rectification,temps_cumuile_rectification)
print("average_retaining et temps_cumuile_rentrainement",average_retaining,temps_cumuile_rentrainement)
```

```python
# Calculation of medians for each position for size, precision, and time for retraining
# Retraining
import statistics

# Create lists to store medians for each position
median_size,median_precision,median_depth,median_num_different_instances= [],[],[],[]
# Calculate the median for each position (size)
for i in range(max(len(sublist) for sublist in number_of_nodes_after_retraining_for_each_rule)):
    position_values = [sublist[i] for sublist in number_of_nodes_after_retraining_for_each_rule if i < len(sublist)]
    median_value = int(statistics.median(position_values))
    median_size.append(median_value)

# Calculate the median for each position (precision)
for i in range(max(len(sublist) for sublist in accuracy_after_retraining_for_each_rule)):
    position_values = [sublist[i] for sublist in accuracy_after_retraining_for_each_rule if i < len(sublist)]
    median_value = statistics.median(position_values)
    median_precision.append(median_value)

# Calculate the median for each position
for i in range(max(len(sublist) for sublist in depth_after_retraining_for_each_rule)):
    position_values = [sublist[i] for sublist in depth_after_retraining_for_each_rule if i < len(sublist)]
    median_value = int(statistics.median(position_values))-1 #To exclude the root from the depth calculation.
    median_depth.append(median_value)

# Calculate the median for each position (number of different instances)
for i in range(max(len(sublist) for sublist in number_of_different_instances_retraining)):
    position_values = [sublist[i] for sublist in number_of_different_instances_retraining if i < len(sublist)]
    median_value = int(statistics.median(position_values))
    median_num_different_instances.append(median_value)

# Calculate the length of instances and its median
number_of_corrections_for_each_tree = [len(i) for i in number_of_different_instances_retraining]
median_number_of_corrections_for_all_trees = int(statistics.median(number_of_corrections_for_each_tree))

print("median_precision retraining",median_precision)
print("median_size retraining",median_size)
print("median_depth retraining",median_depth)
print("median_num_different_instances retraining",median_num_different_instances)
print("median_number_of_corrections_for_all_trees",median_number_of_corrections_for_all_trees)
print("number_of_corrections_for_each_tree",number_of_corrections_for_each_tree)
```

```python
import matplotlib.pyplot as plt

# Create lists to store medians for each position
medians_pre,medians_precisions = [],[]
# Determine the number of positions
#nb_positions = max(len(sublist) for sublist in pre)
nb_positions = median_number_of_corrections_for_all_trees+1

# Create a figure
fig, ax = plt.subplots(figsize=(12, 8))

# Iterate over each position
for i in range(nb_positions):
    # Retrieve data for pre and precisions for this position
    data_pre = [sublist[i] for sublist in accuracy_after_rectification_for_each_rule if i < len(sublist)]
    data_precisions = [sublist[i] for sublist in accuracy_after_retraining_for_each_rule if i < len(sublist)]

    # Add data to the list of medians
    medians_pre.append(data_pre)
    medians_precisions.append(data_precisions)

# Create a boxplot for pre data without outliers
boxplots_pre = ax.boxplot(medians_pre, vert=True, patch_artist=True, positions=[p - 0.25 for p in range(nb_positions)], widths=0.15, showfliers=False)

# Create a boxplot for precisions data without outliers
boxplots_precisions = ax.boxplot(medians_precisions, vert=True, patch_artist=True, positions=[p for p in range(nb_positions)], widths=0.15, showfliers=False)

# Set the color of all pre boxplots to light blue and precisions to red
for box in boxplots_pre['boxes']:
    box.set_facecolor('lightblue')

for box in boxplots_precisions['boxes']:
    box.set_facecolor('red')

# Create a list of intervals [0,1], [1,2], [2,3], ...

# Add axis labels and a title
ax.set_xlabel('Number of rules')
ax.set_ylabel('Accuracy')

# By this line:
ax.set_xticks([p for p in range(0, nb_positions, 2)])
ax.set_xticklabels([f'{i}' for i in range(0, nb_positions, 2)])

# Legend to distinguish the two sets of boxplots
ax.legend([boxplots_pre["boxes"][0], boxplots_precisions["boxes"][0]], ['Rectification', 'Retraining'])

# Set the median line color to black
for line in boxplots_pre['medians']:
    line.set_color('black')

for line in boxplots_precisions['medians']:
    line.set_color('white')

plt.savefig(dataset + '_precision.png', format='png', dpi=300)

# Adjust the x-axis scale to increase the distance between positions
ax.set_xlim()

# Display the plot
plt.show()
```
